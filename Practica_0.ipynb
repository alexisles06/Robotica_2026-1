## Marco teórico

En el sistema ROS2, los nodos son elementos de software que cumplen funciones específicas dentro del entorno robótico. Cada nodo se encarga de ejecutar una tarea determinada, como procesar datos de sensores, controlar actuadores o realizar cálculos, contribuyendo así al funcionamiento general del sistema.

La comunicación entre nodos se realiza a través de los tópicos, que actúan como canales de intercambio de información. Un nodo puede publicar mensajes en un tópico o suscribirse a él para recibir los datos enviados por otros nodos. Este mecanismo permite una comunicación flexible y modular entre diferentes componentes del sistema. Existen diferentes tipos de relaciones en esta comunicación, como one-to-one (uno a uno), one-to-many (uno a varios) y many-to-many (varios a varios), dependiendo de la cantidad de nodos que publiquen o se suscriban a un mismo tópico. Para que la comunicación sea exitosa, es fundamental que tanto el publicador como el suscriptor utilicen el mismo tipo de mensaje, de manera que los datos puedan interpretarse correctamente en ambos extremos del intercambio.

En ROS2, los paquetes son las unidades fundamentales de organización del código. Cada paquete contiene los archivos necesarios para desarrollar una funcionalidad específica dentro del sistema, como nodos, mensajes, configuraciones o archivos de lanzamiento. Al crear un paquete, es importante documentar su estructura para comprender cómo se organizan los archivos y directorios que lo conforman. Esta documentación suele representarse mediante un árbol de directorios, el cual muestra la jerarquía de carpetas y archivos del paquete.

Por ejemplo, la estructura básica de un paquete en ROS2 puede representarse de la siguiente manera:

CMakeLists.txt y package.xml son archivos esenciales que definen las dependencias, el nombre del paquete y la configuración de compilación. La carpeta src/ contiene los scripts o archivos fuente donde se implementan los nodos. El directorio launch/ incluye los archivos de lanzamiento, que permiten ejecutar varios nodos de manera conjunta. Otros directorios, como resource/ o config/, pueden incluir archivos adicionales según las necesidades del proyecto.

Documentar la estructura en forma de árbol facilita la comprensión del proyecto, permite mantener un orden lógico en los archivos y ayuda a otros desarrolladores a identificar rápidamente dónde se ubica cada componente del paquete.

ROS2 cuenta con tres interfaces principales de comunicación entre nodos: Temas (Topics), Servicios (Services) y Acciones (Actions). Cada una de ellas se utiliza según el tipo de interacción o flujo de datos que se requiera dentro del sistema robótico.

- Temas (Topics):
Son canales de comunicación asíncronos mediante los cuales los nodos pueden publicar o suscribirse a mensajes de manera continua. Esta interfaz es ideal para el envío constante de información, como los datos de sensores, el estado de un robot o las lecturas de telemetría. Los tópicos permiten relaciones de tipo one-to-one, one-to-many o many-to-many, facilitando la comunicación entre múltiples nodos de forma eficiente.

- Servicios (Services):
Los servicios permiten una comunicación sincrónica, es decir, de solicitud y respuesta. Un nodo actúa como cliente que realiza una petición, mientras que otro nodo funge como servidor, procesando la solicitud y enviando una respuesta. Esta interfaz es apropiada para tareas que requieren una acción puntual o un resultado inmediato, como cambiar un parámetro, iniciar un proceso o consultar el estado de un componente.

- Acciones (Actions):
Las acciones combinan características de los temas y los servicios, y se utilizan para operaciones que toman tiempo en completarse. Una acción permite enviar una meta (objetivo) a un servidor, recibir retroalimentación sobre su progreso y obtener un resultado final. Esta interfaz es útil, por ejemplo, para mover un robot a una posición determinada, ejecutar trayectorias o realizar tareas prolongadas donde es importante conocer el avance.

Cada una de estas interfaces ofrece ventajas específicas según el tipo de comunicación que se busque establecer entre los nodos, permitiendo diseñar sistemas modulares, eficientes y adaptados a las necesidades del proyecto.

La práctica consiste en el desarrollo de dos nodos, un publicador que comparte una señal senoidal que varía de 1 a -1 constantemente y representa la velocidad angular de un motor en revoluciones por minuto. Así mismo, se crea un nodo subscriptor que recibe la primer señal y la transforma en rad/s con un factor de conversión. Los nodos se presentan a continuación:

```python
#!/usr/bin/env python3

import rclpy
from rclpy.node import Node
from math import sin
from std_msgs.msg import Float64

class MyPublish(Node):
    def __init__(self):
        super().__init__('publish_node')
        self.number_ = 0.0
        self.suma = 0.0
        self.aumento_=0.1
        self.publisher_ =self.create_publisher(Float64, 'publish_topic', 10)
        self.get_logger().info('Nodo publicador activo')
        self.create_timer(0.5,self.subcallback)

    def subcallback(self):
        msg = Float64()
        self.suma += self.aumento_
        msg.data = sin(self.suma)
        self.publisher_.publish(msg)
    
def main(args=None):
    rclpy.init(args=args)
    node = MyPublish()
    rclpy.spin(node)
    rclpy.shutdown()

    if __name__ == "__main__":
        main()
```
